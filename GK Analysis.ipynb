```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("MLS/all_goalkeepers.csv")
df.head()
```

```python
len(df)
```

```python
# gk with highest Sv%
perfect_gk = df[df["Sv%"] == df["Sv%"].max()]
perfect_gk
```

```python
# find best goal keeper out of the perfect ones

best_perfect = perfect_gk[perfect_gk["SV"] == perfect_gk["SV"].max()]
best_perfect
```

```python
# per season averages
per_year = df.groupby("Year").mean(numeric_only=True)
per_year
```

```python
# per club averages
per_club = df.groupby("Club").mean(numeric_only=True)
per_club
```

```python
# find gk above average
mean_saves = df["SV"].mean()
above_average = df[df["SV"] > mean_saves]
above_average
```

```python
# find highest saves per year
max_per_year = df.groupby("Year").apply(lambda x: x.loc[x["SV"].idxmax()])
max_per_year
```

```python
# Find gk who led the league in saves the most times
max_per_player = max_per_year.groupby("Player").count().sort_values(by=["Club"], ascending=False).head(1)
max_per_player
```

Matt Reis has led the MLS in saves for 3 seasons, the most in history

```python
import matplotlib.pyplot as plt

# Group the original dataframe by year and sum the saves for each year
saves_distribution = df.groupby('Year')['SV'].sum()

# Plot the distribution of saves over the years
saves_distribution.plot(kind='bar', xlabel='Year', ylabel='Total Saves', title='Distribution of Saves Over the Years')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = df['SV']
y = df['SHTS']
z = df['W%']

ax.scatter(x, y, z)

ax.set_xlabel('Saves')
ax.set_ylabel('Goals Against Average')
ax.set_zlabel('Win Percentage')

plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = df['Year']
y = df['SV']
z = df['GAA']

ax.plot_trisurf(x, y, z, cmap='viridis', edgecolor='none')

ax.set_xlabel('Year')
ax.set_ylabel('Saves')
ax.set_zlabel('Goals Against Average')

plt.show()
```

```python
# historical average save %
mean_save_pct = df["Sv%"].mean()
mean_save_pct
```

```python
# gk above average
above_average = df[df["Sv%"] >= mean_save_pct]

#gk belove average
below_average = df[df["Sv%"] < mean_save_pct]
```

```python
# highest of the above average
above_average.sort_values(by="Sv%", ascending=False).loc[above_average["SV"] == above_average["SV"].max()]
```

```python
# highest of the belove average
below_average.sort_values(by="Sv%", ascending=False).loc[below_average["SV"] == below_average["SV"].max()]
```
